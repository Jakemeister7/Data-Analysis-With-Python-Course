Pandas - Series

```python
# Import Pandas and Numpy
import pandas as pd
import numpy as np
```

```python
# Create Pandas Series From List (Column in Table)
my_list = [5, 9, 12, 27]
# Create a Series
my_series = pd.Series(my_list)
```

```python
# print without formatting
print(my_series)
```

```python
# print with formatting
my_series
```

```python
# Pull Item From Series with Index Number
print(my_series[1])
my_series[1]
```

```python
# Create Labels with Index Argument
my_index = ['A', 'B', 'C', 'D']
my_series2 = pd.Series(my_list, index=my_index)
my_series2
```

```python
# Create Labels Manually
my_series3 = pd.Series(my_list, ['E', 'F', 'G', 'H'])
my_series3
```

```python
# Pull Item From Series With Label
# print(my_series3.iloc[3])
print(my_series3['G'])
```
