```python
import numpy as np
```

```python
# Sort Numerically
np1 = np.array([6, 7, 4, 9, 0, 2, 1])
np.sort(np1)
print(np1)
print(np.sort(np1))
```

```python
# Reverse?
np9 = np.sort(np1)
print(np9) # ascending low to high
# Reverse it to descending high to low
np9 = np9[::-1]
print(np9)
```

```python
# Sort Alphabetically
np2 = np.array(['Jo', 'Bob', 'Rob', 'Jake', 'Sally'])
print(np2)
# Sort low to high
print(np.sort(np2))
```

```python
# Sort Booleans T(1)/F(0)
np3 = np.array([True, False, False, True])
print(np3)
print(np.sort(np3))
```

```python
# Sort on 2-D Arrays; sorts in each row
np4 = np.array([[6, 7, 1, 9], [8, 3, 5, 0]])
print(np.sort(np4))
```
