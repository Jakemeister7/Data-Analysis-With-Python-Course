Pandas - Changing and Resetting Row Indexes

```python
import pandas as pd
import numpy as np
```

```python
# Import CSV File
url = 'https://raw.githubusercontent.com/flatplanet/Data_Analysis_With_Python_Course/refs/heads/main/dog_data_short.csv'
my_df = pd.read_csv(url)
```

```python
# Change and Reset Row Index
my_df
```

```python
# Create a new column
my_df['Frame Header'] = ['Dog 1', 'Dog 2', 'Dog 3', 'Dog 4', 'Dog 5']
my_df
```

```python
# Set Index
my_df.set_index('Frame Header')
```

```python
# Look - it's gone!
my_df
```

```python
# Make it permanent with inplace
my_df.set_index('Frame Header', inplace=True)
my_df
```

```python
# Reset it
my_df.reset_index()
```

```python
# Not permanent
my_df
```

```python
# Reset Permanent
my_df.reset_index(inplace=True)
```

```python
my_df
```

```python
# delete the old column
my_df.drop('Frame Header', axis=1, inplace=True)
```
