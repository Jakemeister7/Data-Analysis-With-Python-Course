Panda - DataFrame Pulling Data

```python
import pandas as pd
import numpy as np
```

```python
# import from csv from URL
url = 'https://raw.githubusercontent.com/flatplanet/Data_Analysis_With_Python_Course/refs/heads/main/dog_data.csv'
my_df = pd.read_csv(url)
```

```python
my_df
```

```python
# Pull out a row
my_df.loc[0]
```

```python
# Pull out multiple rows 0 and 5
my_df.loc[[0,5]]
```

```python
# Grab First 5 rows
my_df.head(5)
```

```python
# Grab certain number of first rows
my_df.head(9)
```

```python
# Grab the last 5 rows
my_df.tail(5)
```

```python
# Grab a certain number of last rows
my_df.tail(10)
```

```python
# Get Columns using bracket notation(returns panda series)
my_df['DogName']
```

```python
# Get Columns using Dot Notation
my_df.Color
```

```python
# Select specific column using location
my_df.iloc[:, 0]
```

```python
# Select More specific column using location
my_df.iloc[:, 1]
```

```python
# Get information about dataframe
my_df.info()
```

```python
# Get shape of rows and columns
my_df.shape
```

```python
# Get number of dimensions
my_df.ndim
```

```python
# get column datatypes
my_df.dtypes
```

```python
# Get some statistics about the data
my_df.describe()
```

```python
# Describe a specific column - Breed
my_df['Breed'].describe()
```

```python
# Describe a specific column - Color
my_df['Color'].describe()
```

```python
# Describe a specific column - DogName
my_df['DogName'].describe()
```

```python
# Count Data In Dataframe - Count Distinct Values - Descending
my_df['Color'].value_counts()
```

```python
# Count Distinct Values - Ascending
my_df['Color'].value_counts(ascending=True)
```

```python
# Dog Names - without NaN
my_df['DogName'].value_counts()
```

```python
# Dog Names - with NaN
my_df['DogName'].value_counts(dropna=False)
```

```python
# Get relative Frequency - Percentage
my_df['Color'].value_counts(normalize=True)
```

```python
# Get Specific Item Count (Get all white dogs)
my_df['Color'].value_counts()['WHITE']
```

```python
# Count Unique Values - Size
my_df.groupby('Color').size()
```

```python
# Count Unique Values - Count
my_df.groupby('Color').count()
```
