Pandas - Conditionals

```python
import pandas as pd
import numpy as np
```

```python
# import from csv from URL
url = 'https://raw.githubusercontent.com/flatplanet/Data_Analysis_With_Python_Course/refs/heads/main/dog_data.csv'
my_df = pd.read_csv(url)
```

```python
my_df
```

```python
# Conditional Selection < = > <=  >= == etc
my_df == 'BROWN' # Returns a Boolean T/F
```

```python
# Return actual data by passing that into dataframe
my_df[my_df == 'BROWN']
```

```python
# Just return the color column
my_df[my_df == 'BROWN']['Color']
```

```python
# Return Multiple columns
my_df[my_df == 'BROWN'][['Color', 'DogName']]
```

```python
# return a dataframe of the specified info
my_df[my_df['Color'] == 'BROWN']
```

```python
# Multiple Conditional AND &
# my_df[my_df['Color'] == "BROWN"]
my_df[(my_df['Color'] == "BROWN") & (my_df['Breed'] == "MIXED")]
```

```python
# Get the length of that
len(my_df[(my_df['Color'] == "BROWN") & (my_df['Breed'] == "MIXED")])
```

```python
# Multiple Conditional OR
my_df[(my_df['Color'] == "BROWN") | (my_df['Breed'] == "MIXED")]
```

```python
# Length of that
len(my_df[(my_df['Color'] == "BROWN") | (my_df['Breed'] == "MIXED")])
```

```python
# Pull a specific thing from that
my_df[(my_df['Color'] == "BROWN") | (my_df['Breed'] == "MIXED")]['DogName']
```
